```python
df = pd.read_csv(GLP_PATH, sep='\t', header=17, low_memory=False)
```

```python
miRNA_homo_sapiens_df = df[df['Species Scientific Name']=='Homo sapiens']
```

## Matrix

```python
SERIES_MATRIX_PATH = 'GSE159033/GSE159033_series_matrix.txt/GSE159033_series_matrix.txt'
```

```python
series_df = pd.read_csv(SERIES_MATRIX_PATH, sep='\t', header=66, low_memory=False)
series_df = series_df.drop(index=6631)
series_df
```

## Concatenate

```python
miRNA_homo_sapiens_df['ID'] = miRNA_homo_sapiens_df['ID'].astype(str)
series_df['ID_REF'] = series_df['ID_REF'].astype(str)

merged_df = miRNA_homo_sapiens_df.merge(series_df, left_on='ID', right_on='ID_REF', how='inner')
merged_df
```

```python
merged_df.columns
```

```python
merged_df.to_csv('GSE159033_genomic_matrix_matrix.csv')
```
