```python
MATRIX_PATH = 'GSE173787_series_matrix.txt/GSE173787_series_matrix.txt'
```

```python
headers = []
values = []
with open(MATRIX_PATH) as f:
    for i, line in enumerate(f):
        if i >= 27 and i < 64:
            headers.append(line.split('!')[1].split('\t')[0])
            values.append(line.split('!')[1].split('\t')[1:])
```

```python
values_array = np.array(values)
```

```python
# dataframe with columns as headers and rows as columns

df = pd.DataFrame(columns=headers)
for i, row in enumerate(values_array.T):
    df.loc[i] = row

df.head(2)
```
